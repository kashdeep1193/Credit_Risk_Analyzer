```python
def AMT_ANNUITY(df):
  if df['AMT_ANNUITY'] < 20000:
    return 'Low'
  elif df['AMT_ANNUITY'] > 20000 and df['AMT_ANNUITY'] < 50000:
    return 'Medium'
  elif df['AMT_ANNUITY'] > 50000 and df['AMT_ANNUITY'] < 100000:
    return 'High'
  else:
    return 'VeryHigh'
```

```python
df['AMT_ANNUITY_CATE'] = df.apply(AMT_ANNUITY, axis=1)
```
